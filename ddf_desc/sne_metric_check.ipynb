```python
from rubin_sim.maf import ddfBatch
import matplotlib.pylab as plt
%matplotlib inline
import healpy as hp
import rubin_sim.maf.db as db
import rubin_sim.maf.metricBundles as mb
```

```python
file = 'ddf_deep_rolling_slf0.20_v2.2_10yrs.db'

run_name = file.replace('.db', '')
resultsDb = db.ResultsDb(outDir=run_name + "_ddf")


bd = ddfBatch(runName=run_name)
```

```python
final_dict = {}
for key in bd:
    if 'SNN' in key:
        if 'COSMOS' in key:
            final_dict[key] = bd[key]
```

```python
final_dict
```

```python
group = mb.MetricBundleGroup(final_dict,file , outDir=run_name + "_ddf", resultsDb=resultsDb, saveEarly=False)
group.runAll()
group.plotAll(closefigs=False)
```
