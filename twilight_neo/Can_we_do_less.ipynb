```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
import pandas as pd
import sqlite3
```

```python
sql = 'select sunAlt,observationStartMJD from observations where visitExposureTime < 20;'
con = sqlite3.connect('twi_neo_repeat3_z_np2_v2.2_10yrs.db')
df = pd.read_sql(sql, con)
con.close()
```

```python
vat = pd.read_csv('twi_neo_repeat3_riz_np2_v2.2_10yrs__vatiras_granvik_10k_obs.txt', comment='#', delimiter=' ')
```

```python
indx = np.in1d(df['observationStartMJD'], vat['observationStartMJD'])
```

```python
df['sunAlt'].values[indx]
```

```python
vat
```

```python
good = np.where(vat['visitExposureTime'] < 20.)[0]
vat = vat.loc[good]
```

```python
vat
```

```python
mjds = np.unique(vat['time'])
```

```python
mjds
```

```python
sas = np.interp(mjds, df['observationStartMJD'], df['sunAlt'])
```

```python
fig, ax = plt.subplots()
_ack = ax.hist(sas, bins=20)
ax.set_title('twi_neo_repeat3_riz_np2_v2.2')
ax.set_xlabel('Sun Altitude (degrees)')
ax.set_ylabel('N Vatiras observations\nin 15s visits')
```
