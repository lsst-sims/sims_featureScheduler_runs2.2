```python
import numpy as np
import healpy as hp
import astropy.units as u
import rubin_sim.maf.metrics as metrics
import rubin_sim.maf.slicers as slicers
import rubin_sim.maf.plots as plots
import rubin_sim.maf.maps as maps
import rubin_sim.maf.metricBundles as mb
from rubin_sim.maf.batches import scienceRadarBatch
import rubin_sim.maf as maf
import matplotlib.pylab as plt
%matplotlib inline
import copy
```

```python
filename = 'baseline_v2.2_10yrs.db'
```

```python
sb = scienceRadarBatch(runName=filename.replace('.db',''))
```

```python
queries = ['night < %i' % (365.25*i) for i in np.arange(1,11)]
```

```python
key1 = 'baseline_v2_2_10yrs_Parallax_Uncert_@_22_4_HEAL'
key2 = 'baseline_v2_2_10yrs_Proper_Motion_Uncert_@_20_5_HEAL'
```

```python
bl = []
for i,q in enumerate(queries):
    for key in [key1, key2]:
        bundle = copy.deepcopy(sb[key])
        bundle.constraint = q
        bundle.metric.name += ' year %i' % (i+1)
        bundle.fileRoot  += ' year %i' % (i+1)
        bl.append(bundle)
    
```

```python
bg = mb.MetricBundleGroup(bl, filename, outDir='temp')
bg.runAll()
```

```python
bg.plotAll(closefigs=False)
```
