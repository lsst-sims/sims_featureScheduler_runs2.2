```python


import numpy as np
import healpy as hp
import rubin_sim.maf.metrics as metrics
import rubin_sim.maf.slicers as slicers
import rubin_sim.maf.plots as plots
import rubin_sim.maf.maps as maps
import rubin_sim.maf.metricBundles as mb
import rubin_sim.maf as maf

import matplotlib.pylab as plt
%matplotlib inline
```

```python
outDir='temp'
resultsDb = maf.db.ResultsDb(outDir=outDir)

db_file = 'noroll_v2.2_10yrs.db'

bundleList = []

metric = metrics.PassMetric(cols=['night'])
slicer = slicers.UserPointsSlicer(ra=[20,20,180,180], dec=[0, -20, 0, -20])
bundle = mb.MetricBundle(metric,slicer,"")

bundleList.append(bundle)
bg = maf.metricBundles.MetricBundleGroup(bundleList, db_file, outDir=outDir, resultsDb=resultsDb)
bg.runAll()


for point in bundleList[0].metricValues:
    nights = np.sort(point['night'])
    num = np.arange(nights.size)
    plt.plot(nights, num)
plt.xlabel('Night')
plt.ylabel('N obs')

plt.vlines(np.arange(1,11)*365.25, ymin=-20, ymax=900, ls='--', color='black', alpha=0.5)
```
