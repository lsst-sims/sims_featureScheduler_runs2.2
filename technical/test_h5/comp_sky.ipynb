```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
import pandas as pd
import sqlite3
```

```python
con = sqlite3.connect('test_h5_v2.2_10yrs.db')
sql = 'select filter,skyBrightness,fiveSigmaDepth from observations;'
new = pd.read_sql(sql, con)
con.close()

con = sqlite3.connect('../../baseline/baseline_v2.2_10yrs.db')
baseline = pd.read_sql(sql, con)
con.close()
```

```python
filters = 'ugrizy'


```

```python
for filtername in filters:
    
    fig,ax = plt.subplots()
    good =np.where(baseline['filter'].values == filtername)[0]
    _tmp, bins, _p = ax.hist(baseline['skyBrightness'].values[good], bins=100, alpha=0.5)
    good =np.where(new['filter'].values == filtername)[0]

    _tmp, bins, _p  = ax.hist(new['skyBrightness'].values[good], bins=bins, alpha=0.5)

    ax.set_title(filtername)
    ax.set_xlabel('sky brightness (mag/sq arcsec)')
```

```python
for filtername in filters:
    
    fig,ax = plt.subplots()
    good =np.where(baseline['filter'].values == filtername)[0]
    _tmp, bins, _p = ax.hist(baseline['fiveSigmaDepth'].values[good], bins=100, alpha=0.5)
    good =np.where(new['filter'].values == filtername)[0]

    _tmp, bins, _p  = ax.hist(new['fiveSigmaDepth'].values[good], bins=bins, alpha=0.5)

    ax.set_title(filtername)
    ax.set_xlabel('5-sigma depth (mag)')
```
